```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt




edges = data2[['NomeRegioneValidato', 'distretto']].dropna()

# bipartite graph
B = nx.Graph()


regions = edges['NomeRegioneValidato'].unique()
districts = edges['distretto'].unique()

B.add_nodes_from(regions, bipartite=0, type='region')
B.add_nodes_from(districts, bipartite=1, type='district')

# weghting the edges
edge_weights = edges.groupby(['NomeRegioneValidato', 'distretto']).size().reset_index(name='weight')

for _, row in edge_weights.iterrows():
    B.add_edge(row['NomeRegioneValidato'], row['distretto'], weight=row['weight'])

# adding graphs (italia mafia confiscated related only!)
plt.figure(figsize=(12, 10))

pos = {}
pos.update((node, (1, i)) for i, node in enumerate(regions))   # x=1 for regions
pos.update((node, (2, i)) for i, node in enumerate(districts)) # x=2 for districts


nx.draw_networkx_nodes(B, pos, nodelist=regions, node_color='lightblue', node_size=800, label="Regions")
nx.draw_networkx_nodes(B, pos, nodelist=districts, node_color='lightgreen', node_size=800, label="Districts")


edges_to_draw = nx.draw_networkx_edges(B, pos, alpha=0.5,
                                       width=[d['weight']*0.1 for (u,v,d) in B.edges(data=True)])

nx.draw_networkx_labels(B, pos, font_size=8)

plt.title("Bipartite Network: Mafia Confiscated Assets (Regions <--> Districts)")
plt.legend(scatterpoints=1)
plt.axis("off")
plt.show()

from networkx.algorithms import bipartite

region_nodes = [n for n, d in B.nodes(data=True) if d["type"] == "region"]
R = bipartite.weighted_projected_graph(B, region_nodes)

plt.figure(figsize=(10, 8))
pos = nx.spring_layout(R, k=0.5, seed=42)


nx.draw_networkx_nodes(R, pos, node_color='lightblue', node_size=1200)


weights = [d['weight'] for (_,_,d) in R.edges(data=True)]
nx.draw_networkx_edges(R, pos, width=[w*0.1 for w in weights], alpha=0.4)


nx.draw_networkx_labels(R, pos, font_size=9)

plt.title("Region–Region Mafia Asset Network (via District Overlaps)")
plt.axis("off")
plt.show()
```

```python
import pandas as pd
import networkx as nx

# using centrality measures
deg_cent = nx.degree_centrality(R)

#strength to be added!
weighted_deg = dict(R.degree(weight="weight"))

bet_cent = nx.betweenness_centrality(R, weight="weight", normalized=True)


centrality_df = pd.DataFrame({
    "DegreeCentrality": deg_cent,
    "WeightedDegree": weighted_deg,
    "BetweennessCentrality": bet_cent
}).sort_values("WeightedDegree", ascending=False)

# Top 10 regions
centrality_df.head(10)
```
